```python
conta = "PM10"

model_pm_24 = joblib.load(str('./entrenamiento_sin_persistencia/modelos/'+ conta +'/mal/'+ conta + '_'+str(24)+'hora.pkl'))

data_hour_merge_PM10 = pd.read_csv("/DATA/paw/jupyterhub_notebook/daniel.bustillos/DanielBustillos/contaminaciÃ³n/pipeline-norberto/recover/datos_modelos_PM10.csv")

target = "PM10mean_max_frcst_"

lista = []
for i in range(1,25):
    lista.append(str(target+str(i)))

hora = 24

dataset_PM10 = data_hour_merge_PM10.copy()
#dataset.drop(["fecha"],axis=1,inplace=True)
dataset_PM10.dropna(inplace=True)

X_PM10 = dataset_PM10.loc[:, dataset_PM10.columns != str(target+str(hora))]
y_PM10 = dataset_PM10[str(target+str(hora))]

X_PM10.drop(lista[hora:],axis=1,inplace=True)
if hora != 1:
    X_PM10.drop(lista[:hora-1],axis=1,inplace=True)

#Definimos los conjuntos de prueba:

X_train_PM10, X_test_PM10, y_train_PM10, y_test_PM10 = train_test_split(X_PM10, y_PM10, test_size=0.3, random_state=123)

#Predict 
pronos_PM10 = model_pm_24.predict(X_test_PM10)
```
