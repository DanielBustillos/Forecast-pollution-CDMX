```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None 
```

```python
data_hour_merge_24 = pd.read_csv('./datos_entrenamiento.csv')
```

### Saquemos las componentes del viento:

```python
def componente_x(magnitud, direccion):
    comp_x = magnitud*np.cos(direccion)
    comp_y = magnitud*np.sin(direccion)
    lista = [comp_x,comp_y]
    return lista
```

```python
data_hour_merge_24['WSP_x'] = 0
data_hour_merge_24['WSP_y'] = 0

lst = data_hour_merge_24.apply(lambda x: componente_x(x['WSP'], x['WDR']), axis=1)

data_hour_merge_24[['WSP_x']] = [item[0] for item in lst]
data_hour_merge_24[['WSP_y']] = [item[1] for item in lst]
```

```python
data_hour_merge_24.head()
```

### Arreglemos las fechas:

```python
data_hour_merge_24 = data_hour_merge_24[(data_hour_merge_24.fecha > "2015-01-01 23:00:00")]
data_hour_merge_24['dia'] = data_hour_merge_24['dia'].astype(str).astype(float)
data_hour_merge_24['mes'] = data_hour_merge_24['mes'].astype(str).astype(float)
```

Creemos un diccionario con __data_hour_merge_24__ dividido por estación:

```python
estaciones = data_hour_merge_24.id_station.unique().tolist()

data_est = {}
for elem in estaciones:
    data_est[elem] = data_hour_merge_24[data_hour_merge_24.id_station == elem]
```
